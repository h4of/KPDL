```python
def standardiztion(text):
  text = text.lower()
  text = re.sub(r'<.*?>', '', text)             # xóa html tag
  text = re.sub(r'http\S+|www\S+', '', text)    # link
  text = re.sub(r'\S+@\S+', '', text)           # email
  text = re.sub(r'\d+', '', text)               # số
  text = re.sub(r'[^\w\s]', ' ', text)          # ký tự đặc biệt
  return text
```

```python
def tokenization(text):
  return underthesea.word_tokenize(text)
```

```python
def remove_stopwords(text):
  return [word for word in text if word not in stopwords]
```

```python
def convert_no_diacritics_to_diacritics(text):
    text = ' '.join(text)
    for k, v in mapping.items():
        text = text.replace(k, v)
    return text
```

```python
sens ="Đây là một cuốn sách rất hay và tôi đã đọc nó hai lần."
print(tokenization(sens))
```

```python
dmt_comments['comment'] = dmt_comments['comment'].apply(standardiztion).apply(tokenization).apply(remove_stopwords).apply(convert_no_diacritics_to_diacritics)
dmt_comments['comment']
```
